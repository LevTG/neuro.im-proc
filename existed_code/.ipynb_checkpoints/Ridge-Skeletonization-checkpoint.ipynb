```python
def percentile_rescale(arr, plow=1, phigh=99):
    vmin,vmax = np.percentile(arr, (plow, phigh))
    if vmin == vmax:
        return np.zeros_like(arr)
    else:
        return np.clip((arr-vmin)/(vmax-vmin),0,1)
    
```

```python
def threshold_obj_size(mask, min_size=3):
    out = np.zeros(mask.shape, np.bool)
    labels, nl = ndi.label(mask)
    objs = ndi.find_objects(labels)
    for i,o in enumerate(objs):
        submask = labels[o]==i+1
        sz = np.sum(submask)
        if sz >= min_size:
            out[o][submask] = True
    return out

def sato2d_ms(img, scales):
    return np.max([hcecp.sato2d(img, scale)*scale**2 for scale in scales], axis=0)

import ucats
def noise_assisted_sato(img, sigmas, noise_sigma=None, threshold=0.1, niters=100, noise_scaling=1):
    out = np.zeros_like(img)
    if noise_sigma is None:
        noise_sigma = ucats.mad_std(img)
        print('estimated noise sigma:', noise_sigma)
    for i in tqdm(range(niters)):
        img_n = img + randn(*img.shape)*noise_sigma*noise_scaling
        #img_n = ucats.scramble.local_jitter2d(img,1,1)
        #sx = skfilters.sato(img_n, sigmas, black_ridges=False, mode='constant')
        #sx = np.max([sato2d(img, s)*s**2 for s in sigmas], axis=0)
        sx = sato2d_ms(img_n, sigmas)
        out += threshold_obj_size(sx > threshold)
        #out += sx > threshold
        #out += sx
    return out/niters
```

```python
from imfun.core import extrema

def peaks_in_rows(arr, mask):
    mx_rows = [np.array([r[0] for r in extrema.locextr(arr[k], refine=False, output='max',sort_values=False) 
                        if mask[k,r[0]]], dtype=int)
                 for k in (range(arr.shape[0]))]
    out_mask = np.zeros(arr.shape,bool)
    for k,mx in enumerate(mx_rows):
        if len(mx):
            out_mask[k,mx[mx<arr.shape[1]]] = True
    return out_mask
    
    

def rotated_peaks(arr,mask,angle=45):
    "Row-wise local maxima in image after rotation by angle"
    r_arr = ndi.rotate(arr, angle, order=1, mode='nearest')
    r_mask = ndi.rotate(mask, angle, order=1).astype(bool)
    peaks = peaks_in_rows(r_arr, r_mask).astype(float32)
    rot_mask = ndi.rotate(peaks, -angle, order=1)
    sh = arr.shape
    rot_sh = rot_mask.shape
    crop = int(floor((rot_mask.shape[0] - sh[0])/2)) # this has to be fixed for non-rectangular images
    
    if crop > 0:
        rot_mask = rot_mask[crop:-crop-1,crop:-crop-1]
    #print(arr.shape, crop, rot_sh, rot_mask.shape)
    return rot_mask > 0.25

def simple_oriented_peaks(arr, mask, sigma=3):
    if sigma > 0:
        proj_s = ndi.gaussian_filter(arr, sigma)
    else:
        proj_s = arr
    #mx_rows_mask = peaks_in_rows(proj_s, mask).astype(int)
    #mx_cols_mask = peaks_in_rows(proj_s.T, mask.T).T.astype(int)
    angles = (0, 90, -3*45/2, -45, -45/2, 45/2, 45, 3*45/2)
    mx_rots = [rotated_peaks(proj_s, mask, angle).astype(int) for angle in angles]
    
    #return np.sum([mx_rows_mask, mx_cols_mask] + mx_rots,axis=0)
    return mx_rots
```

```python
import numba

@numba.jit
def make_trajectory(x0, force_field, xtrails, speed_trails, food_field, vizit_counts,
                    food_memory=3,
                    ignore_transitions=False,
                    m=1, friction=0.1, T=100, dt=1,tol=0.001):
    traj = np.zeros((T,) + (len(x0),))
    #print(traj.shape)
    
    nr,nc,nd = force_field.shape
    
    F = force_field
    
    food_cargo = np.zeros(food_memory)
    food_idx = 0
    food_acc = 0
    food_count = 0
    ri_p, ci_p = -1,-1

    traj[0] = x0
    for i in range(0, T-1):
        r,c = traj[i]
        ri,ci = np.int(r), np.int(c)
        if ri >= nr or ci >= nc or r < 0 or c < 0:
            break
            
        if ignore_transitions | (ri != ri_p) | (ci != ci_p):
            food_new = food_field[ri,ci] 
            food_idx = (food_idx+1)%food_memory # point to next position
            food_tail = food_cargo[food_idx] # either zero or end of ring
            food_cargo[food_idx-1] = food_new
        
            food_count = min(food_count+1, food_memory) # visited or length of ring
            food_acc += food_new - food_tail
                
            if food_count >= food_memory:
                xtrails[ri,ci] += food_acc/food_count
                vizit_counts[ri,ci] += 1
            food_count += 1            
            
            
        a = F[ri,ci]/m 
        if i == 0:
            next_state = traj[i] + a*dt**2/2
        else:
            next_state = (2-friction)*traj[i] - (1-friction)*traj[i-1] + a*dt**2

        speed = np.sum((next_state - traj[i])**2)**0.5
        
        if speed < tol:
            break
        speed_trails[ri,ci] = speed_trails[ri,ci] + speed 

        traj[i+1] = next_state
        ri_p,ci_p = ri,ci
    return traj[:i]

@numba.jit#(parallel=True)
def collect_trails(field, 
                   food_field,
                   mask=None, 
                   m=10, 
                   food_memory=3,
                   mag_th=0.25, dt=0.25,friction=0.05, T=1000):
    sh = field.shape[:-1]
    xtrails = np.zeros(sh)
    vtrails = np.zeros(sh) 
    counts = np.zeros(sh)
    
    mag = np.sum(field**2,axis=-1)**0.5
    if mask is None:
        mask = mag > mag_th
        
    for r in range(sh[0]):
        for c in range(sh[1]):
            if mask[r,c]:
                make_trajectory(np.array([r,c]), 
                                field, 
                                xtrails, 
                                vtrails, 
                                food_field,
                                counts,
                                food_memory=food_memory,
                                m=m,dt=dt,friction=friction,T=T)
    return xtrails, vtrails,counts
```

```python
def mean_pair(x1,x2):
    return (x1+x2)/2



def turbosnail_vesselness(img, sigma, amp=1, min_counts=1, **kwargs):
    #H = hessian_by_dog(img, sigma)
    #H = hcecp.skf.hessian_matrix(img_noisy,sigma)
    #lam,v = hcecp.hessian_eigen_decomp(H)
    #mask = (lam[...,0] < 0) & ((lam[...,1] <= 0) | (lam[...,1] < np.abs(lam[...,0])/0.5))
    #weights = percentile_rescale(np.abs(lam[...,0]))*mask
    #field = v[...,:,1]*weights[...,np.newaxis]
    sato,Vf = hcecp.sato2d(img, sigma, hessian_variant='gradient_of_smoothed', return_vectors=True)
    
    Vf = Vf[...,0][...,::-1] #?
    #Vf = Vf[...,0,:][...,::-1] #?
    weights = hcecp.percentile_rescale(sato)
    
    field = amp*Vf*weights[...,None]
    mask= np.ones(img.shape, bool)
    
    xtrails_fw, vtrails_fw, counts_fw = collect_trails(field,img,mask=mask, **kwargs) 
    xtrails_bw, vtrails_bw, counts_bw = collect_trails(-field,img,mask=mask,**kwargs)
    
    counts = mean_pair(counts_fw,counts_bw)
    xtrails = np.where(counts >= min_counts, mean_pair(xtrails_fw, xtrails_bw)/(1e-6+counts), 0)
    vtrails = np.where(counts >= min_counts, mean_pair(vtrails_fw, vtrails_bw), 0)
    
    return xtrails, vtrails, counts
```

```python
def dog(m,k,sigma):
    ndim = np.ndim(m)
    trunc = 6 # default
    if np.any(sigma*trunc < 3):
        trunc = 3/sigma
    o = np.zeros(ndim, np.int)
    o[k] = 1
    return ndi.gaussian_filter(m, sigma, order=o, truncate=trunc)
```
